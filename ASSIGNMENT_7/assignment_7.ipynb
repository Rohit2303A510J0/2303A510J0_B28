#Question 1

```python
import pandas as pd

# Create sample CO2 emissions dataset
data = {
    "Year": [1960, 1970, 1980, 1990, 2000, 2010, 2020],
    "CO2_Emissions": [9400, 14500, 19000, 22000, 25000, 31000, 34000]  # in million metric tons
}

# Save as CSV
df = pd.DataFrame(data)
df.to_csv("co2_emissions.csv", index=False)

print("CSV file created successfully!")
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("co2_emissions.csv")

# Plot Line Chart
plt.figure(figsize=(10,6))
plt.plot(df["Year"], df["CO2_Emissions"], marker="o", linestyle="-", color="b")

# Labels and title
plt.title("Global CO₂ Emissions Over Time", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("CO₂ Emissions (Million Metric Tons)", fontsize=12)
plt.grid(True)

# Show plot
plt.show()
```

```python
# Check growth from first to last year
start = df["CO2_Emissions"].iloc[0]
end = df["CO2_Emissions"].iloc[-1]
change = ((end - start) / start) * 100

if end > start:
    trend = f"CO₂ emissions increased by {change:.2f}% from {df['Year'].iloc[0]} to {df['Year'].iloc[-1]}."
else:
    trend = f"CO₂ emissions decreased by {abs(change):.2f}% from {df['Year'].iloc[0]} to {df['Year'].iloc[-1]}."

print(trend)
```

#Question - 2

```python
import pandas as pd

# Create sample Student Performance dataset
data = {
    "Exam": [1, 2, 3, 4, 5, 6],
    "Score": [55, 65, 70, 75, 80, 88]  # hypothetical scores
}

# Save as CSV
df = pd.DataFrame(data)
df.to_csv("student_scores.csv", index=False)

print("CSV file created successfully!")
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("student_scores.csv")

# Plot Line Chart
plt.figure(figsize=(8,5))
plt.plot(df["Exam"], df["Score"], marker="o", linestyle="-", color="g")

# Labels and Title
plt.title("Student Performance Over Exams", fontsize=16)
plt.xlabel("Exam Number", fontsize=12)
plt.ylabel("Score (%)", fontsize=12)
plt.ylim(0, 100)  # set scale 0-100 for scores
plt.grid(True)

# Show plot
plt.show()
```

```python
start = df["Score"].iloc[0]
end = df["Score"].iloc[-1]
change = end - start

if end > start:
    trend = f"Student’s performance improved by {change} points from Exam {df['Exam'].iloc[0]} to Exam {df['Exam'].iloc[-1]}."
else:
    trend = f"Student’s performance declined by {abs(change)} points over the exams."

avg_improvement = df["Score"].diff().mean()

print(trend)
print(f"On average, the student improved by {avg_improvement:.2f} points per exam.")
```
