#Question-1

```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------
# Step 1: Create synthetic dataset (traffic_source.csv)
# ---------------------------
data = {
    "Source": ["Direct", "Search", "Social", "Referral", "Ads"],
    "Visits": [1500, 3200, 1200, 800, 2200]  # synthetic values
}

df = pd.DataFrame(data)
df.to_csv("traffic_source.csv", index=False)

# ---------------------------
# Step 2: Load dataset
# ---------------------------
traffic_df = pd.read_csv("traffic_source.csv")

# ---------------------------
# Step 3: Plot bar chart
# ---------------------------
plt.figure(figsize=(8,6))
bars = plt.bar(traffic_df["Source"], traffic_df["Visits"],
               color=["skyblue","lightgreen","salmon","orange","violet"])

# Add value labels above bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, str(yval),
             ha='center', va='bottom', fontsize=10)

plt.title("Website Traffic by Source", fontsize=14)
plt.xlabel("Traffic Source")
plt.ylabel("Number of Visits")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

# ---------------------------
# Step 4: Insights
# ---------------------------
highest_source = traffic_df.loc[traffic_df["Visits"].idxmax()]
print(f"Highest traffic comes from {highest_source['Source']} with {highest_source['Visits']} visits.")
```

#Question - 2

```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------
# Step 1: Create synthetic dataset (monthly_rainfall.csv)
# ---------------------------
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
              "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Rainfall": [45, 50, 70, 90, 120, 200, 300, 280, 180, 100, 60, 50]  # synthetic values in mm
}

df = pd.DataFrame(data)
df.to_csv("monthly_rainfall.csv", index=False)

# ---------------------------
# Step 2: Load dataset
# ---------------------------
rain_df = pd.read_csv("monthly_rainfall.csv")

# ---------------------------
# Step 3: Define seasonal colors
# ---------------------------
season_colors = {
    "Winter": "skyblue",   # Dec, Jan, Feb
    "Spring": "lightgreen",# Mar, Apr, May
    "Summer": "orange",    # Jun, Jul, Aug
    "Autumn": "salmon"     # Sep, Oct, Nov
}

month_to_season = {
    "Jan": "Winter", "Feb": "Winter", "Dec": "Winter",
    "Mar": "Spring", "Apr": "Spring", "May": "Spring",
    "Jun": "Summer", "Jul": "Summer", "Aug": "Summer",
    "Sep": "Autumn", "Oct": "Autumn", "Nov": "Autumn"
}

colors = [season_colors[month_to_season[m]] for m in rain_df["Month"]]

# ---------------------------
# Step 4: Plot bar chart
# ---------------------------
plt.figure(figsize=(10,6))
bars = plt.bar(rain_df["Month"], rain_df["Rainfall"], color=colors)

# Add value labels above bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, str(yval),
             ha='center', va='bottom', fontsize=9)

plt.title("Average Monthly Rainfall (mm)", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

# ---------------------------
# Step 5: Insights
# ---------------------------
wettest = rain_df.loc[rain_df["Rainfall"].idxmax()]
driest = rain_df.loc[rain_df["Rainfall"].idxmin()]

print(f"Wettest month: {wettest['Month']} ({wettest['Rainfall']} mm).")
print(f"Driest month: {driest['Month']} ({driest['Rainfall']} mm).")
```
