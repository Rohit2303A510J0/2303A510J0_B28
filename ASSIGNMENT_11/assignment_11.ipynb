# Question 1

```python
import matplotlib.pyplot as plt

sources = ["Direct", "Search", "Social", "Referral", "Ads"]
visits = [4200, 6800, 3100, 2500, 1500]

plt.figure(figsize=(8, 5))
bars = plt.bar(sources, visits)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        str(height),
        ha='center',
        va='bottom'
    )

plt.xlabel("Traffic Source")
plt.ylabel("Number of Visits")
plt.title("Website Traffic by Source")
plt.tight_layout()
plt.show()
```

# Question 2

```python
import matplotlib.pyplot as plt

# Sample rainfall data (since no CSV is available)
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

rainfall = [12, 20, 35, 50, 80, 150, 220, 200, 120, 60, 30, 15]

# Season-based colors
colors = [
    "lightblue", "lightblue",        # Winter (Jan, Feb)
    "lightgreen", "lightgreen",      # Spring (Mar, Apr)
    "orange", "orange",              # Summer (May, Jun)
    "blue", "blue", "blue",          # Monsoon (Jul, Aug, Sep)
    "brown", "brown", "brown"        # Autumn/Winter (Oct, Nov, Dec)
]

plt.figure(figsize=(10, 5))
bars = plt.bar(months, rainfall, color=colors)

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        str(height),
        ha="center",
        va="bottom"
    )

plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.title("Average Monthly Rainfall")
plt.tight_layout()
plt.show()
```
