#Question 1

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(42)
data = pd.DataFrame({
    "Student": np.arange(1, 101),
    "Subject": np.random.choice(["Math", "Science", "English", "History"], 100),
    "Gender": np.random.choice(["Male", "Female"], 100),
    "Score": np.random.normal(loc=70, scale=10, size=100).clip(0, 100)
})

sns.set_theme(style="whitegrid", palette="muted", font_scale=1.1)
cat_plot = sns.catplot(
    data=data,
    x="Subject",
    y="Score",
    hue="Gender",
    kind="box",
    height=5,
    aspect=1.4,
    palette="coolwarm"
)
cat_plot.set_axis_labels("Subject", "Exam Score (%)")
cat_plot.set_titles("Exam Results by Subject and Gender")
cat_plot.fig.suptitle("Categorical Plot: Customized Appearance", fontsize=14, y=1.05)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

sns.set_theme(style="darkgrid", palette="pastel")
sns.catplot(
    data=data,
    x="Subject",
    y="Score",
    hue="Gender",
    kind="strip",
    dodge=True,
    height=5,
    aspect=1.4
).set(title="Exam Results (Darkgrid Theme)")
plt.show()
```

#Question 2

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
data = pd.read_csv(url)

corr_matrix = data.corr()

sns.set_theme(style="white", font_scale=1.1)

plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
    cbar_kws={"shrink": 0.8}
)

plt.title("Correlation Matrix - Boston Housing Data", fontsize=14, pad=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
